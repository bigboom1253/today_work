```python
target_ls = ['스포츠', '카페', '여행', '패션', '맛집', '반려동물','게임','미디어감상']
```

```python
import json

file_path = "./sample_plus_target.json"

json_data = []

for idx in range(len(post_label)) :
    json_data.append({
        "user_id_ls": post_label['user_id_ls'][idx],
        "post_date_ls": post_label['post_date_ls'][idx],
        "contents_ls": post_label['contents_ls'][idx],
        "img_url_ls": post_label['img_url_ls'][idx],
        "url_ls" : post_label['url_ls'][idx],
        "hashtag" : post_label['hashtag'][idx],
        "target" : post_label['target'][idx]
    })

with open(file_path, 'w') as outfile:
    json.dump(json_data, outfile)
```
