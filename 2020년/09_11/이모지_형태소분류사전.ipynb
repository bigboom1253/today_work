```python
emoji.UNICODE_EMOJI
```

```python
emoji_morpheme_dic = {}
for key, value in emoji.UNICODE_EMOJI.items():
    emoji_morpheme_dic[key] = key + '_' + value
```

```python
emoji_morpheme_dic
```

# 유의사항

- emoji를 품사태깅하는 dictionary를 파일로 저장하고 쓰게되면, 업데이트 된 이모티콘은 반영할 수 없음.(emoji 패키지는 계속 업데이트가 되기 때문에)
- 하지만, 계속 위와 같은 코드로 이모지 처리를 할때마다 새로운 사전을 구축하는 방법도 탐탁치않음.
