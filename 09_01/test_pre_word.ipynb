```python
import json

contents_ls = []
img_url_ls = []
url_ls = []
post_date_ls = []
user_id_ls = []

for i in range(1, 78) : 
    with open("{}.json".format(i), "r") as fp:
        temp_json =json.load(fp)
        print('{}'.format(i))
        for source_idx in range(len(temp_json)) :
            try :
                contents_ls.append(temp_json[source_idx]['content'])
                img_url_ls.append(temp_json[source_idx]['image_url'])
                post_date_ls.append(temp_json[source_idx]['post_date'])
                user_id_ls.append(temp_json[source_idx]['insta_id'])
                url_ls.append(temp_json[source_idx]['url'])
            except :
                continue

contents_ls
```

```python
import re
import MeCab
from konlpy.tag import Komoran


class Pre():

    # 문장 토큰화, Escape Code 있는 상태에서 - \n 기준으로 나눔
    def sent_tokenize(self, datas):
        return [d.split('\n') for d in datas]

    # Escape Code 처리
    escape_code = ['\n', '\xa0', '\"', '\'', '\t', '\r', '\$', '\\', '\u200d']
    def del_escape(self, datas):
        for e in self.escape_code:
            datas = datas.replace(e, ' ')
        return datas

    # 한글 단어 토큰화 (단어 길이 1일 경우 제거)
    def word_tokenize(self, datas):
        if type(datas) != list:
            datas = [datas]
        return list(filter(lambda i : i if len(i) > 1 else False, self.search('(?<!가-힣)[가-힣]+(?!가-힣)', datas)))

    # 불용어 처리
    def get_stop(self):
        with open ('kor_stopwords.txt', 'r') as f:
            stopwords = f.readlines()
        return list(map(lambda i : i.replace('\n', ''), stopwords))

    def del_stopword(self, token_datas):
        if type(token_datas) != list:
            token_datas = [token_datas]

        return [list(filter(lambda i : True if i not in self.get_stop() else False, d)) for d in token_datas]

    # 정규표현식 Functions (finditer, sub) - return = list
    def search(self, reg, datas):
        if type(datas) != list:
            datas = [datas]
        result = [list(re.finditer(reg, d)) for d in datas]
        features = []
        for r in result:
            features.extend(list(map(lambda i : i.group(), r)))
        return features
    
    def substr(self, reg, datas):
        if type(datas) != list:
            datas = [datas]
        return [re.sub(reg, ' ', d) for d in datas]

    # POS Tagging
    # Mecab 1순위, KOMORAN 2순위 사용 고려
    # 안쓰는 품사는 제거하거나 쓸 품사만 가져오기
    def pos_tag(self, datas, pack = 'mecab'):
        if pack == 'mecab':
            # Mecab
            mecab = mecab.MeCab()
            return [mecab.pos(d) for d in datas]

        elif pack=='komoran':
            # KOMORAN
            komoran = Komoran()
            return [komoran.pos(d) for d in datas]

        else:
            return

    # Tokenizing

# Ngram화
```

```python
pre_word = Pre()

content_data = []
for content in contents_ls :
     content_data.append(pre_word.del_escape(content))

content_data = pre_word.sent_tokenize(content_data)

data = []

for i in content_data :
    data.append(pre_word.word_tokenize(i))
data    
```

```python
from chatspace import ChatSpace

spacer = ChatSpace()
data2 = []
count = 0
for i in data :
    if count%1000 == 0 :
        print('{}개 진행중'.format(count))
    count += 1
    try :
        temp = []
        for j in spacer.space(i, batch_size=128) :
            for word in j.split() :
                temp.append(word)
        data2.append(temp)

    except :
        continue
data2
```

```python
import json

file_path = "./sample.json"

json_data = []

count_idx = 1
for idx in range(len(data2)) :
    json_data.append({
        "user_id_ls": user_id_ls[idx],
        "post_date_ls": post_date_ls[idx],
        "contents_ls": data2[idx],
        "img_url_ls": img_url_ls[idx],
        "url_ls" : url_ls[idx]
    })

with open(file_path, 'w') as outfile:
    json.dump(json_data, outfile)
```
